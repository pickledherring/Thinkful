```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from sklearn.naive_bayes import BernoulliNB
```

```python
df = pd.read_csv('data/amazon_cells_labelled.txt', delimiter= '\t', header=None)
df.columns = ['message', 'sentiment']
keywords = ['great', 'highly recommend', 'excellent', 'must have', 'best', 'awesome']

for keyword in keywords:
    df[keyword] = df.message.str.contains(' ' + keyword + ' ', case=False)
```

```python
sns.heatmap(df.corr())
plt.show()
```

```python
data = df[keywords]
target = df['sentiment']
```

```python
bnb = BernoulliNB()
bnb.fit(data, target)
pred = bnb.predict(data)

print('{} out of {} points misclassified'.format((target != pred).sum(), data.shape[0]))
```

Works okay, but it has no negative sentiment features.
