```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

Create a "purchases" dataframe and plot ad views against items purchased. There doesn't look to be any relationship between the two variables.

```python
purchases = pd.read_csv('data/purchases.csv')
plt.scatter(x=purchases['ad_views'], y=purchases['items_purchased'], color='brown', marker='x', s=20)
plt.show()
```

A side-by-side comparison of items purchased and ad views. The first has a more normal distribution, while the second is skewed far to the lower side.

```python
plt.figure(figsize=(5, 3))
           
plt.subplot(1, 2, 1)
plt.boxplot(purchases['items_purchased'])
           
plt.subplot(1, 2, 2)
plt.boxplot(purchases['ad_views'])

plt.tight_layout()
plt.show
```

Displays the frequency of purchases and ad views in certain ranges. The range of ads seen is apparently much larger than the range of items purchased. The total number of items purchased in Canada could be small enough to be inadequately representative.

```python
plt.hist(purchases['ad_views'], color='red', alpha=.5)
plt.hist(purchases['items_purchased'], color='blue', alpha=.5)
plt.show()
```

The two bar graphs below separate ad views and purchases by country, showing that the US buys and views more than Canada in this dataset.

```python
purchases.groupby('country').ad_views.sum().plot(kind='bar', title='Ad views by country')
plt.show
```

```python
purchases.groupby('country').items_purchased.sum().plot(kind='bar', title='Items purchased by country')
plt.show
```
