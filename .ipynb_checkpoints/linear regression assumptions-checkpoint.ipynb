```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sqlalchemy import create_engine
import seaborn as sns

%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

import warnings
warnings.filterwarnings(action="ignore")

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

df = pd.read_sql_query('select * from weatherinszeged',con=engine)

engine.dispose()
```

Build a regression model where the target variable is temperature. As explanatory variables, use humidity, windspeed, windbearing, and pressure. Estimate the model using OLS.

```python
Y = df.temperature
X = df[['humidity', 'windspeed', 'windbearing', 'pressure']]
X = sm.add_constant(X)

results = sm.OLS(Y, X).fit()
results.summary()
```

Now, check if your model meets the Gauss-Markov Conditions above. If some of the assumptions are not met, discuss the implications of the violations for the correctness of your model.

The model should account for a linear relationship between the features and the target, and the constant should account for the error term.

```python
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
predictions = lrm.predict(X)
errors = Y - predictions

print("Mean of the errors in the medical costs model is: {}".format(np.mean(errors)))
```

```python
plt.scatter(predictions, errors)
plt.show()
```

The errors increase greatly in range with the predictions and swing downwards. Some predictive feature is missing that explains the error change or outliers are affecting the data. This will throw the confidence interval off for the observations that have larger error values.

```python
from scipy.stats import bartlett
from scipy.stats import levene

print(bartlett(predictions, errors))
print(levene(predictions, errors))
```

```python
sns.heatmap(df[['humidity', 'windspeed', 'windbearing', 'pressure', 'temperature']].corr())
```

Humidity and temperature are highly correlated. This will cause some features' weights to be misestimated.

```python
plt.plot(errors)
plt.show()

from statsmodels.tsa.stattools import acf

acf_data = acf(errors)

plt.plot(acf_data[1:])
plt.show()
```

The errors are correlated with each other. A relevant feature is missing that is reflected in the errors.

```python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
df = pd.read_sql_query('select * from houseprices',con=engine)
engine.dispose()
```

```python
df.drop(['id'], axis=1, inplace=True)
```

```python
## Fill continuous variable null values with zero
for column in ['masvnrarea', 'lotfrontage', 'garagecars']:
    df[column] = df[column].fillna(0)
    
indexes = df[df.garageyrblt == 'None'].index
df.loc[indexes, ['garageyrblt']] = 1980
df['garageyrblt'] = pd.to_numeric(df['garageyrblt'])

## Fill all null values with 'none'
df = df.fillna('None')
```

```python
df2 = df[['yearbuilt', 'yearremodadd', 'masvnrarea', 'bsmtfinsf1', 'fireplaces', 'garagecars',
          'wooddecksf', 'openporchsf', 'totalbsmtsf', 'firstflrsf', 'secondflrsf','fullbath',
          'saleprice']]
```

```python
Y = df2.saleprice
X = df2.drop(columns = ['saleprice'])
lrm = linear_model.LinearRegression()
lrm.fit(X, Y)
predictions = lrm.predict(X)
errors = Y - predictions
```

```python
print('errors mean: {}'.format(np.mean(errors)))

plt.scatter(predictions, errors)
plt.title('errors vs. predictions')
plt.show()

print('bartlett: {}'.format(bartlett(predictions, errors)))
print('levene: {}'.format(levene(predictions, errors)))

sns.heatmap(df2.corr())
plt.show()

plt.plot(errors)
plt.title('errors')
plt.show()

acf_data = acf(errors)
plt.plot(acf_data[1:])
plt.show()
```

From what I can see above, the features and target should have a linear relationship and the mean error is sufficiently low. There are a few extreme errors probably caused by large outliers, and error variance does change over the instances. Features have low correlation with other features, and errors are not correlated with each other. Error outliers appear to be a major problem and should be eliminated, and other features to explain some of the error variance change might be needed.
