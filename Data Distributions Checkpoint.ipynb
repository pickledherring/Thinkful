```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
binomial = np.random.binomial(200, .3, 100)
```

```python
plt.hist(binomial)
plt.show()
```

```python
plt.hist(binomial)
plt.axvline(binomial.mean(), color='g')
plt.axvline(binomial.mean() - binomial.std(), color='g', linestyle='dotted')
plt.axvline(binomial.mean() + binomial.std(), color='g', linestyle='dotted')
plt.show()
```

The binomial distribution seems suited to being described by the mean and standard deviation. The binomial and the normal distributions are not skewed, so standard deviation accurately describes variability.

```python
a = np.random.normal(5, .5, 100)
b = np.random.normal(10, 1, 100)
c = a + b
```

```python
plt.hist(c)
plt.show
```

```python
plt.hist(c)
plt.axvline(c.mean(), color='g')
plt.axvline(c.mean() - c.std(), color='g', linestyle='dotted', linewidth=2)
plt.axvline(c.mean() + c.std(), color='g', linestyle='dotted', linewidth=2)
plt.show()
```

The mean does appear to be in the center, but the data is so oddly lumped that the mode might be more useful. standard deviation does still providde some useful information to a limited extent.
